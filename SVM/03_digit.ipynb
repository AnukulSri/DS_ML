```python
df = pd.DataFrame(ld['data'],ld['target'])
```

```python
df['target'] = ld.target
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3)
```

```python
from sklearn.svm import SVC
```

```python
vm = SVC()
```

```python
vm.fit(X_train,Y_train)
```

```python
pred= vm.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(Y_test,pred))
```

```python
print(classification_report(Y_test,pred))
```

```python
vm.score(X_test,Y_test)
```

# Grid Search

```python
param_grid={'C':[0,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001],'kernel':['linear','rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gsv = GridSearchCV(SVC(),param_grid,refit=True,verbose=2)
```

```python
gsv.fit(X_train,Y_train)
```

```python
prediction = gsv.predict(X_test)
```

```python
print(confusion_matrix(Y_test,prediction))
```

```python
print(classification_report(Y_test,prediction))
```

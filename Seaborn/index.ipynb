# Distribution Plot

- this create a histogram 
- in this we have to pass the dataframe frame name and column name

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.distplot(df['total_bill'])# distplot means distribution plot
```

# Jointplot
#### in this there is an argument - kind

```python
sns.jointplot(x='total_bill',y='tip',data=df)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg') # reg means regression a machine learning concept
```

# Pair plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex') # in hue we have to pass a categorical column means non numeric column
```

# Rugplot 
- it draws a dash on the graph

```python
sns.rugplot(df['total_bill'])
```

# kdeplot (Kernel density estimation)
- it is represents in distplot

```python
sns.kdeplot(df['total_bill'])
```

# Categorical Plot

- Categorical means non numeric column 

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=df,estimator=np.std)
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,hue='smoker')
```

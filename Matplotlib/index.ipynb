# Matplotlib

- It is the most popular plotting library for python
- It gives you to control over every aspect of a figure
- It was designed to have a similar feel to MatLab's graphicall plotting
- It was created by john hunter

#### Installation : pip install matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
# this is to see the plots which we create.
%matplotlib inline 
```

```python
x = np.linspace(0,11,11)
y = x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) # the 1 no. is row number 2 is column number 3 is plot number
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,'g-')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'r-')
axes[1].plot(y,x,'g-')
```

## Figure Size and Dpi

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,5))
axes[0].plot(x,y)
axes[1].plot(y,x)
#plt.savefig('mat_pic.png',dpi=200) # to save the plot as a image
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X square')
ax.plot(x,x**3,label= 'X Cube')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,linestyle=':',marker='o',markersize=8)
```
